# How will we access our data?

The typical way that files are accessed within code is through *file paths*.

You may have noticed when you are browsing through files something that looks like:  
**/pictures/vacation/2025-06-22_001.jpg**

This references a `.jpg` file in the Folder (or directory) called **vacation**, and that folder (or directory) is inside another folder called **pictures**.

We are going to play a game called [bashcrawl](https://gitlab.com/slackermedia/bashcrawl) that will help us learn about paths now.

### About bashcrawl

bashcrawl is a game that let's us learn how to use the terminal. `bash` is a shell scripting language that let's you interact with files and directories using the terminal. Our goal is to learn about *file paths*, and so our specific approach is to build a map.

### How to setup bashcrawl

1. Install [Cygwin](https://cygwin.com/install.html). Download the file using the `setup-x86_64.exe` link on the linked page. Run the file that is downloaded and click `Next` through the steps.
*Note: You should select `install from the internet` option and you will need to select a `mirror` on one of the screens*

1. One the final screen of the installation, keep the `Add icon to Start Menu` checkbox selected.

1. Download bashcrawl using the instructions on this [page](https://gitlab.com/slackermedia/bashcrawl).

1. Open `Cygwin`.

1. Open your `Downloads` folder (or wherever you saved the bashcrawl zip file).

1. Right-click on the zip file and select `Extract All`, then click on the `Extract` button. This will create a new folder containing the files for the game. The default should open the new folder.

1. Then open the only folder there, it should look like the image below.  


1. From the bashcrawl folder, right-click on the folder labeled entrance and click on `Copy as path`.

1. In Cygwin type `cd` and then `space` and then right-click and select `paste`. You should see something like the image below.  


1. We're almost there. In Cygwin, delete the double quotes (`"`), change `C:` to `/cygdrive/c` and change all `\` to `/`. An example is shown below.  

*Note: I understand that this seems tedious, but it is the first step in helping you to know that the `/` is important for what we will be doing and how to access paths.

1. Finally, press the `Enter` key and type `cat scroll` and then `Enter` again. You are now playing the game.
